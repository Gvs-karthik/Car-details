```python
# Use LabelEncoder to handle categorical data

from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()

claim['INCOME'] = LE.fit_transform(claim['INCOME'])
claim['PARENT1'] = LE.fit_transform(claim['PARENT1'])
claim['HOME_VAL'] = LE.fit_transform(claim['HOME_VAL'])
claim['MSTATUS'] = LE.fit_transform(claim['MSTATUS'])
claim['GENDER'] = LE.fit_transform(claim['GENDER'])
claim['EDUCATION'] = LE.fit_transform(claim['EDUCATION'])
claim['OCCUPATION'] = LE.fit_transform(claim['OCCUPATION'])
claim['CAR_USE'] = LE.fit_transform(claim['BLUEBOOK'])
claim['BLUEBOOK'] = LE.fit_transform(claim['CAR_USE'])
claim['CAR_TYPE'] = LE.fit_transform(claim['CAR_TYPE'])
claim['RED_CAR'] = LE.fit_transform(claim['RED_CAR'])
claim['OLDCLAIM'] = LE.fit_transform(claim['OLDCLAIM'])
claim['REVOKED'] = LE.fit_transform(claim['REVOKED'])
claim['CLM_AMT'] = LE.fit_transform(claim['CLM_AMT'])
claim['URBANICITY'] = LE.fit_transform(claim['URBANICITY'])
claim['BIRTH'] = LE.fit_transform(claim['BIRTH'])
claim['AGE'] = LE.fit_transform(claim['AGE'])
claim['YOJ'] = LE.fit_transform(claim['YOJ'])
claim['CAR_AGE'] = LE.fit_transform(claim['CAR_AGE'])
```

```python
IndepVar = []
for col in claim.columns:
    if col != 'CLM_AMT':
        IndepVar.append(col)

TargetVar = ('CLM_AMT')

x = claim[IndepVar]
y = claim[TargetVar]
```

```python
# Split the data into train and test sets

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Display dataset information

claim.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

mmscaler = MinMaxScaler(feature_range=(0, 1))

x_train = mmscaler.fit_transform(x_train)
x_train = pd.DataFrame(x_train)

x_test = mmscaler.fit_transform(x_test)
x_test = pd.DataFrame(x_test)
```

```python
x_test
```

# Linear Regression Algorithm

```python
# Train the algorithm and build the model with train dataset

from sklearn.linear_model import LinearRegression

# Create model object

MulRGR = LinearRegression()

#LinearRegression(fit_intercept=True, normalize='deprecated', copy_X=True, n_jobs=None, positive=False)

# Train the model with training dataset (70%)

MulRGR.fit(x_train, y_train)
```

```python
# Predict the model with test dataset

y_pred = MulRGR.predict(x_test)
```

```python
# Evaluation metrics for Regression analysis

from sklearn import metrics

print('Mean Absolute Error (MAE):', round(metrics.mean_absolute_error(y_test, y_pred),3))  
print('Mean Squared Error (MSE):', round(metrics.mean_squared_error(y_test, y_pred),3))  
print('Root Mean Squared Error (RMSE):', round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),3))
print('R2_score:', round(metrics.r2_score(y_test, y_pred),6))
print('Root Mean Squared Log Error (RMSLE):', round(np.log(np.sqrt(metrics.mean_squared_error(y_test, y_pred))),3))
```

```python
# Define the function to calculate the MAPE - Mean Absolute Percentage Error

def MAPE (y_test, y_pred): 
    y_test, y_pred = np.array(y_test), np.array(y_pred)
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

# Evaluation of MAPE 

result = MAPE(y_test, y_pred)
print('Mean Absolute Percentage Error (MAPE):', round(result, 3), '%')
```
